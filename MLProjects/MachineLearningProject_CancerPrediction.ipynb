```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

#BINARY CLASSIFICATION therefore last layer activation is sigmoid
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(X_train,y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses
```

```python
losses.plot()
```

```python
#PREVENTING OVERFITTING DROPOUT and EARLY_STOP
```

```python
model=Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

#BINARY CLASSIFICATION therefore last layer activation is sigmoid
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(X_train,y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history).plot()
```

```python
from tensorflow.keras.layers import Dropout
#Deopout drops the neurons number
model=Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dropout(rate=0.5))

model.add(Dense(15,activation='relu'))
model.add(Dropout(rate=0.5))

#BINARY CLASSIFICATION therefore last layer activation is sigmoid
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(X_train,y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
losses= pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
#CALSSIFICATION
predict = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predict))
```

```python
print(confusion_matrix(y_test,predict))
```
